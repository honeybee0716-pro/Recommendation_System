```python
import psycopg2
import re
```

```python
conn = psycopg2.connect("dbname='movie_recommender' user='postgres' host='localhost' port=5432 password=''")
```

```python
conn.autocommit = True
```

```python
mov_file = open('../data/movie.csv')
```

```python
def insert(record_to_insert, counter):
    postgres_insert_query = "UPDATE idmapper SET newid='%s' where oldid='%s'"%(record_to_insert)
    cur = conn.cursor()
    cur.execute(postgres_insert_query)
    conn.commit()
```

```python
def fetchNewId(name):
    sql = "select m.tconst from movies as m inner join ratings as r on m.tconst = r.tconst where r.numvotes > 1000 and  m.primarytitle like E'"+ name +"' order by r.numvotes DESC limit 1"
    cur = conn.cursor()
    cur.execute(sql)
    record = None
    for row in cur:
        record = row
    if not record: return None
    return record[0]
    
```

```python
def cleanAndInsert(record, counter):
    name = record[1]
    name = re.sub(r'\([^)]*\)', '', name).replace('"', '').strip()
    name = name.replace("'", "''")
    new_id = fetchNewId(name)
    if not new_id: return
    old_id = str(record[0])
    record_to_insert = (new_id, old_id)
    insert(record_to_insert, counter)
```

```python
counter = 0
record = True
mov_file.next()
while record:
    record = mov_file.next()
    record = record.split(',')
    print counter, record
    cleanAndInsert(record, counter)
    counter += 1
```
